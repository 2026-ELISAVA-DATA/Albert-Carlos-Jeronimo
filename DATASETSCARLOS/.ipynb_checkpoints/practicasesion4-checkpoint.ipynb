```python
import numpy as np
import pandas as pd
df = pd.read_csv("Shoe prices.csv")
df.head(50)
```

```python
Now pandas can treat the column as quantitative data.
```

```python

df["Price (USD)"] = (
    df["Price (USD)"]
    .replace(r"[$,]", "", regex=True)
    .astype(float)
)
```

```python
df = df.sort_values("Price (USD)")
df = df.sort_values("Price (USD)", ascending=False)
```

```python
df.head(40)
```

```python
df_expensive = df[df["Price (USD)"] >= 200]
df_expensive
```
