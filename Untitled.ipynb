```python
import requests
import pandas as pd
import time
import re
from bs4 import BeautifulSoup
from urllib.parse import urlparse, parse_qs
from tqdm import tqdm
HEADERS = {
    "User-Agent": "Mozilla/5.0 (compatible; MACBASkateResearch/1.0)"
}

QUERY = "MACBA skate"
MAX_RESULTS = 2000
RESULTS_PER_PAGE = 30
SLEEP_TIME = 1.5

def duckduckgo_search(query, max_results):
    urls = set()
    page = 0

    while len(urls) < max_results:
        offset = page * RESULTS_PER_PAGE
        search_url = "https://duckduckgo.com/html/"
        params = {
            "q": query,
            "s": offset
        }

        r = requests.post(search_url, data=params, headers=HEADERS, timeout=15)
        soup = BeautifulSoup(r.text, "lxml")

        links = soup.select("a.result__a")
        if not links:
            break

        for link in links:
            href = link.get("href")
            if href:
                parsed = urlparse(href)
                qs = parse_qs(parsed.query)
                if "uddg" in qs:
                    clean_url = qs["uddg"][0]
                    urls.add(clean_url)

        page += 1
        time.sleep(SLEEP_TIME)

    return list(urls)[:max_results]

def extract_date(soup):
    meta_props = [
        ("meta", {"property": "article:published_time"}),
        ("meta", {"name": "pubdate"}),
        ("meta", {"name": "date"}),
        ("meta", {"name": "DC.date.issued"})
    ]

    for tag, attrs in meta_props:
        el = soup.find(tag, attrs=attrs)
        if el and el.get("content"):
            return el["content"][:10]

    text = soup.get_text(" ", strip=True)
    match = re.search(r"\b(20\d{2}[-/]\d{2}[-/]\d{2})\b", text)
    if match:
        return match.group(1)

    return ""

def scrape_page(url):
    try:
        r = requests.get(url, headers=HEADERS, timeout=15)
        soup = BeautifulSoup(r.text, "lxml")

        title = soup.title.string.strip() if soup.title else ""
        date = extract_date(soup)

        if "macba" in r.text.lower() and "skate" in r.text.lower():
            return {
                "Title": title,
                "Date": date,
                "URL": url
            }
    except Exception:
        return None

    return None

print("ğŸ” Searching DuckDuckGo...")
urls = duckduckgo_search(QUERY, MAX_RESULTS)
print(f"Found {len(urls)} candidate URLs")

data = []

for url in tqdm(urls, desc="Scraping pages"):
    result = scrape_page(url)
    if result:
        data.append(result)
    time.sleep(0.5)

df = pd.DataFrame(data).drop_duplicates(subset="URL")
df.to_csv("macba_skate_articles.csv", index=False, encoding="utf-8")

df.head()
```
